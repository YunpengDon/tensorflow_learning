# 1. Gradient decent in Himmelblau function

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

## Create an input grid and its himmelblau output

```python
def himmelblau(x):
    return (np.square(x[0]) + x[1] - 11) ** 2 + (x[0] + np.square(x[1]) -7) ** 2 
```

```python
x = np.arange(-6, 6, 0.1)
y = np.arange(-6, 6, 0.1)
x.shape, y.shape
```

```python
X, Y = np.meshgrid(x, y)
X.shape, Y.shape
```

```python
Z = himmelblau([X, Y])
```

## Draw a graph of the output

```python
fig = plt.figure('himmelblau', figsize=(10, 8))
ax = fig.gca(projection='3d')

ax.plot_surface(X, Y, Z,cmap='GnBu')
ax.view_init(60, -30)
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', fontsize=14)
plt.show()
```

```python
fig = plt.figure(figsize=(6, 6))
ax2 = plt.contour(X, Y, Z, [2 ** x for x in range(0, 10)], cmap='gray')
plt.show()
```

## GD optimization

```python
x = tf.constant([1., 2.])
```

```python
result = []
for step in range(200):
    with tf.GradientTape() as tape:
        tape.watch(x)
        z = himmelblau(x)
    grads = tape.gradient(z, x)
    result.append([x, z])
    if step % 20 ==19:
        print('step {}: x = {}, z={}'.format(step, x.numpy(),z.numpy()))
    x -= 0.01 * grads
```

```python
fig = plt.figure(figsize=(6, 6))
ax2 = plt.contour(X, Y, Z, [2 ** x for x in range(0, 10)], cmap='gray')
count=0
for rslt in result:
    plt.plot(rslt[0][0], rslt[0][1], '+')
    count += 1
plt.show()
```
